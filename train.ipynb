```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
dataset_path = "./dataset/"

train_data = tf.keras.utils.audio_dataset_from_directory(

    dataset_path,
    batch_size=6,
    shuffle=True,
    validation_split=0.2,
    subset="training",
    output_sequence_length=64000,
    ragged=False,
    label_mode="categorical",
    labels="inferred",
    sampling_rate=None,
    seed=1998
)

validation_data = tf.keras.utils.audio_dataset_from_directory(

    dataset_path,
    batch_size=6,
    shuffle=True,
    validation_split=0.2,
    subset="validation",
    output_sequence_length=64000,
    ragged=False,
    label_mode="categorical",
    labels="inferred",
    sampling_rate=None,
    seed=25
)
```

```python
model = tf.keras.models.Sequential([

    tf.keras.layers.Conv1D(64, kernel_size=32, strides=20, activation="relu", input_shape=(64000, 1)),
    tf.keras.layers.MaxPool1D(2),
    tf.keras.layers.Conv1D(32,kernel_size=4, activation="relu"),
    tf.keras.layers.MaxPool1D(4),
    tf.keras.layers.Conv1D(32, kernel_size=4, activation="relu"),
    tf.keras.layers.MaxPool1D(6),
    tf.keras.layers.Conv1D(28, kernel_size=4, activation="relu"),
    tf.keras.layers.MaxPool1D(2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(22, activation="softmax")

])
```

```python
model.compile(

    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=['accuracy']
)
```

```python
model.fit(
    train_data,
    validation_data=validation_data,
    epochs=128
)
```

```python
model.save('weights/AudioClassificatio.h5')
```
